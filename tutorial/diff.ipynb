# Systematic differences

We look at the systematic differences between words.
We do this per language, so: separately for the Pali original and the Latin translation.

The idea is that if the same difference occurs between various word pairs, we have a morphological
relationship between the pair.

We load the corpus.

```python
# A = use("etcbc/dhammapada:clone", checkout="clone", hoist=globals())
A = use("etcbc/dhammapada", hoist=globals())
```

# Latin

We collect the differences between pairs of words.

```python
D = collectDiffs(A, "word", "latin", frequencyThreshold=0, sizeThreshold=4)
```

Here are the top-100 differences.

```python
D.showDiffs(0, 100)
```

# Pali

We collect the differences between pairs of words.

```python
D = collectDiffs(A, "word", "pali", frequencyThreshold=0, sizeThreshold=4)
```

```python
D.showDiffs(0, 100)
```
